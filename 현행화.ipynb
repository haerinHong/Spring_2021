```
# 공통영역에 해당하는 리버스 데이터를 거르기 위한 작업
# filte_rev에 공통영역에 해당하는 리버스 데이터를 넣음. 스키마만 따기 위한 작업
filter_rev = pd.DataFrame(rev[rev['테이블명'] == ''])
filter_rev
```

```
rev_col = rev['테이블명'].unique().tolist()
for i in range(len(rev_col)) :
  rev_col[i] = rev_col[i].upper()

rev_col
```

```
# 두개의 유니크한 테이블 명의 교집합을 찾는다
intersection = list(set(asis_col) & set(rev_col))
print( intersection )
print(len(intersection))
```

```
# asis 공통영역에 해당하는 테이블명 리스트에 해당하는 리버스 테이블명이 있으면 filter_rev에 저장.
for i in range(len(rev.index)) :
  if rev['테이블명'][i].upper() in asis_col:
    filter_rev = pd.concat([filter_rev, rev[i:i+1]], ignore_index=True)

  filter_rev
```
