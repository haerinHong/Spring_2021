```
# 변경) file_asis 위치 -> 해당 주제영역 일괄 편집 후 엑셀 파일로 저장한 것, 이름은 주제영역이 포함되어야함

def filtering(asis_name) :
  asis_name = asis_name
  file_asis = '/gdrive/MyDrive/Colab Notebooks/[ASIS]/{}.xlsx'.format(asis_name)
  asis = pd.read_excel(file_asis)

  # asis의 unique한 테이블명을 list로 저장
  asis_col = asis['테이블명'].unique().tolist()
 # 공통영역에 해당하는 리버스 데이터를 거르기 위한 작업
  # filte_rev에 공통영역에 해당하는 리버스 데이터를 넣음. 스키마만 따기 위한 작업
  filter_rev = pd.DataFrame(rev[rev['테이블명'] == ''])

  # for문
  # 교집합에 해당하는 리버스 데이터를 넣음.
  
  for i in range(len(rev.index)) :
    rev_uppers = rev['테이블명'][i].upper()
    if rev_uppers in asis_col:
      filter_rev = pd.concat([filter_rev, rev[i:i+1]], ignore_index=True)
  
  output_file_path_FILTER = '/gdrive/MyDrive/Colab Notebooks/[현행화]/{}.xlsx'.format(asis_name)
  filter_rev.to_excel(output_file_path_FILTER, index=False)
  sample = pd.read_excel(output_file_path_FILTER)
  sample
```

```
import os
def toKorean(asis_name) :
  menu = asis_name

  # file_path = '/gdrive/MyDrive/Colab Notebooks/[현행화]/{}.xlsx'.format(menu) #현행화 뽑은 것
  output_file_path_FILTER = '/gdrive/MyDrive/Colab Notebooks/[현행화]/{}.xlsx'.format(asis_name)
  file_path = output_file_path_FILTER
  file_name = os.path.basename(file_path)
  file_path2 = '/gdrive/MyDrive/Colab Notebooks/[ASIS]/{}.xlsx'.format(menu) #as_is
  output_file_name = '[한글변경]/한글_'+ file_name

  filter_rev_3 = pd.read_excel(file_path)
  asis = pd.read_excel(file_path2)

  asis_col = asis_col = asis['테이블명'].unique().tolist()
  asis_2 = asis.copy()
  filter_rev_tablename = filter_rev_3['테이블명'].unique()
  asis_2['테이블명'] = asis_2['테이블명'].str.lower()
  asis_2['컬럼명'] = asis_2['컬럼명'].str.lower()

  #원본 엔터티이름을 한글로 바꾸는 작업 -> HashMap 생성
  to_change_entity_filter_rev_tablename = []
  for i in range(len(filter_rev_tablename)) :
    real_entity_name = asis_2[asis_2['테이블명'] == filter_rev_tablename[i]]['원본엔터티명'].unique()[0]
    to_change_entity_filter_rev_tablename.append(real_entity_name)
  change_col1 = {}

  for i in range(len(filter_rev_tablename)) :
    aa = filter_rev_tablename[i]
    bb = to_change_entity_filter_rev_tablename[i]
    change_col1[aa] = bb

  # 엔터티명 일괄 변경 
  for i in range(len(filter_rev_3.index)) :
    filter_rev_3['원본엔터티명'][i] = change_col1.get( filter_rev_3['원본엔터티명'][i])


  # 속성명 한글로 일괄 변경----
  # asis 기준 HashMap 생성
  col_to_kor = {}
  for i in range(len(asis_2.index)) :
    if asis_2['컬럼명'][i] is None:
      continue
    else :
      colname =  asis_2['컬럼명'][i]
      propname = asis_2['속성명'][i]
      if colname not in col_to_kor :
        col_to_kor[colname] = propname

  # 엔터티명 일괄 변경 
  for i in range(len(filter_rev_3.index)) :
    filter_rev_3['속성명'][i] = col_to_kor.get(filter_rev_3['컬럼명'][i])

  # 속성명이 None인 경우 (ASIS에는 없는데 리버스에 있는 경우, 참조할 대상이 없으로 None이 나옴. 이런 경우 기존 컬럼명이랑 동일하게 만듬)
  for i in range(len(filter_rev_3.index)) :
    if filter_rev_3['속성명'][i] == None :
      filter_rev_3['속성명'][i] = filter_rev_3['컬럼명'][i]

  # 파일로 출력
  output_file_name = '[한글변경]/한글_'+ file_name
  output_file_path = '/gdrive/MyDrive/Colab Notebooks/[현행화]/{}.xlsx'.format(output_file_name)
  filter_rev_3.to_excel(output_file_path, index=False)
  sample = pd.read_excel(output_file_path)
  sample
```

## 엔터티명은 같은데 테이블 명이 다른 경우 필터링 체크

```
menu_error = '다국어지원'
file_path2 = '/gdrive/MyDrive/Colab Notebooks/[ASIS]/{}.xlsx'.format(menu_error) #as_is
asis = pd.read_excel(file_path2)
rev = pd.read_excel('/gdrive/MyDrive/Colab Notebooks/리버스_일괄편집.xlsx')
```

```
# 기존 asis 영역에서 원본엔터티명과 테이블명을 하나의 Hashmap 형태로 저장,
asis_enti_table = {}
for i in range(len(asis.index)) :
  entity = asis['원본엔터티명'][i]
  table = asis['테이블명'][i]
  if entity not in asis_enti_table :
    asis_enti_table[entity] = table
asis_enti_table
```

```
# 찾는 것이 없으면 처리해야하는 로직 테스트
test_dic = {}
test_dic['a'] = 'aa'
test_dic['b'] = 'bb'

if test_dic.get('bb') == None :
  print('찾는 대상이 딕셔너리에 없습니다')
```

```
# 원본엔터티명과 리버스파일의 원본엔터티와 비교해 같은 것이 있다면, 그 리버스파일의 원본엔터티-테이블명을 HashMap(딕셔너리) 로 담아서 저장한다.
#  filter_rev_3['원본엔터티명'][i] = change_col1.get(filter_rev_3['원본엔터티명'][i])

same_asis_ent = {}
diff_asis = {}
diff_rev = {}

for i in range(len(rev.index)) :
  rev_entity = rev['원본엔터티명'][i].upper()
  rev_table = rev['테이블명'][i].upper()

  if asis_enti_table.get(rev_entity) == None : # ASIS에는 있는데 변경되어서 리버스에는 없을때
    continue
  else : # ASIS에 리버스의 엔터티명이 있을때
    if asis_enti_table.get(rev_entity) == rev_table : # 정확히 ASIS와 리버스의 (원본엔터티명 - 테이블명)이 일치하는 경우
      if rev_entity not in same_asis_ent:
        same_asis_ent[rev_entity] = rev_table
    else : # 우려하는 상황 <- 원본엔터티명은 같으나 테이블명이 서로 다른 경우
        if rev_entity not in diff_rev:
          diff_rev[rev_entity] = rev_table
          diff_asis[rev_entity] = asis_enti_table.get(rev_entity)
```
